```python
class WrongAge(Exception):
    pass

class AgeInvalid(Exception):
    pass

class Father:
    def __init__(self, age):
        if age < 0:
            raise WrongAge("Age cannot be negative")
        self.age = age

class Son(Father):
    def __init__(self, father_age, son_age):
        super().__init__(father_age)  # Call the constructor of the base class

        if son_age >= father_age:
            raise AgeInvalid("Son's age cannot be greater than or equal to Father's age")

        self.son_age = son_age

# Example usage:
try:
    father = Father(40)
    son = Son(40, 30)  
except WrongAge as e:
    print(f"Error: {e}")
except AgeInvalid as e:
    print(f"Error: {e}")
else:
    print("Both Father and Son are created successfully.")
```

```python
class FormulaError(Exception):
    def __init__(self, message="Invalid formula"):
        self.message = message
        super().__init__(self.message)

def calculate(formula):
    try:
        elements = formula.split()
        
        if len(elements) != 3:
            raise FormulaError("Input must consist of 3 elements")
        
        num1 = float(elements[0])
        operator = elements[1]
        num2 = float(elements[2])

        if operator not in ['+', '-']:
            raise FormulaError("Invalid operator. Use only '+' or '-'")

        if operator == '+':
            result = num1 + num2
        else:
            result = num1 - num2

        print(f"Result: {result}")

    except ValueError:
        raise FormulaError("Invalid input. Numbers must be valid floats")

# Interactive calculator loop
while True:
    user_input = input("Enter a formula (or 'quit' to exit): ")
    
    if user_input.lower() == 'quit':
        break
    
    try:
        calculate(user_input)
    except FormulaError as e:
        print(f"Error: {e}")
```

```python
import threading
import random
import time
import queue

def random_number_generator(queue):
    while True:
        number = random.randint(1, 100)
        print(f"Generated number: {number}")
        queue.put(number)
        time.sleep(1)

def square_calculator(queue):
    while True:
        number = queue.get()
        if number % 2 == 0:
            square = number ** 2
            print(f"Square of {number}: {square}")
        time.sleep(1)

def cube_printer(queue):
    while True:
        number = queue.get()
        if number % 2 != 0:
            cube = number ** 3
            print(f"Cube of {number}: {cube}")
        time.sleep(1)

# Create a shared queue
shared_queue = queue.Queue()

# Create threads
generator_thread = threading.Thread(target=random_number_generator, args=(shared_queue,))
square_thread = threading.Thread(target=square_calculator, args=(shared_queue,))
cube_thread = threading.Thread(target=cube_printer, args=(shared_queue,))

# Start threads
generator_thread.start()
square_thread.start()
cube_thread.start()

# Wait for threads to finish
generator_thread.join()
square_thread.join()
cube_thread.join()
```

```python
import threading
import queue
import time
import random

class VegetableMarket:
    def __init__(self, capacity):
        self.capacity = capacity
        self.vegetables = {'Tomato': 0, 'Carrot': 0, 'Cucumber': 0}
        self.market_queue = queue.Queue()

    def add_vegetables(self, vegetable, quantity):
        if vegetable in self.vegetables:
            self.vegetables[vegetable] += quantity
            print(f"Farmer produced {quantity} {vegetable}(s). Total {vegetable}(s): {self.vegetables[vegetable]}")
        else:
            print(f"Unknown vegetable: {vegetable}")

    def sell_vegetables(self, vegetable, quantity):
        if vegetable in self.vegetables and self.vegetables[vegetable] >= quantity:
            self.vegetables[vegetable] -= quantity
            print(f"Consumer bought {quantity} {vegetable}(s). Remaining {vegetable}(s): {self.vegetables[vegetable]}")
        else:
            print(f"Out of stock for {vegetable}. Consumer is willing to wait.")

    def market_status(self):
        print(f"\nMarket Status:")
        for vegetable, quantity in self.vegetables.items():
            print(f"{vegetable}: {quantity}")
        print(f"Queue Length: {self.market_queue.qsize()}\n")

def farmer_thread(market, vegetable, produce_interval):
    while True:
        time.sleep(produce_interval)
        quantity = random.randint(1, 5)
        market.add_vegetables(vegetable, quantity)

def consumer_thread(market, vegetable, purchase_interval):
    while True:
        time.sleep(purchase_interval)
        quantity = random.randint(1, 3)
        market.sell_vegetables(vegetable, quantity)

def market_thread(market, check_interval):
    while True:
        time.sleep(check_interval)
        market.market_status()

# Create the Vegetable Market
market = VegetableMarket(capacity=10)

# Create Farmer Threads
tomato_farmer = threading.Thread(target=farmer_thread, args=(market, 'Tomato', 4))
carrot_farmer = threading.Thread(target=farmer_thread, args=(market, 'Carrot', 5))
cucumber_farmer = threading.Thread(target=farmer_thread, args=(market, 'Cucumber', 6))

# Create Consumer Threads
tomato_consumer = threading.Thread(target=consumer_thread, args=(market, 'Tomato', 3))
carrot_consumer = threading.Thread(target=consumer_thread, args=(market, 'Carrot', 2))
cucumber_consumer = threading.Thread(target=consumer_thread, args=(market, 'Cucumber', 4))

# Create Market Thread
market_monitor = threading.Thread(target=market_thread, args=(market, 8))

# Start all threads
tomato_farmer.start()
carrot_farmer.start()
cucumber_farmer.start()
tomato_consumer.start()
carrot_consumer.start()
cucumber_consumer.start()
market_monitor.start()

# Wait for all threads to finish (Ctrl+C to interrupt)
try:
    tomato_farmer.join()
    carrot_farmer.join()
    cucumber_farmer.join()
    tomato_consumer.join()
    carrot_consumer.join()
    cucumber_consumer.join()
    market_monitor.join()
except KeyboardInterrupt:
    print("Simulation interrupted.")
```

```python
import threading
import time

# Global variable for synchronization
mutex = threading.Lock()

# Function to print numbers from 1 to 5
def function_A():
    for i in range(1, 6):
        time.sleep(1)  # Simulating time-consuming task
        with mutex:
            print(i, end=" ")
    print()

# Function to print letters A to E
def function_B():
    for char in 'ABCDE':
        time.sleep(1)  # Simulating time-consuming task
        with mutex:
            print(char, end=" ")
    print()

# Create threads
thread_A = threading.Thread(target=function_A)
thread_B = threading.Thread(target=function_B)

# Start threads
thread_A.start()
thread_B.start()

# Wait for both threads to complete
thread_A.join()
thread_B.join()

print("Program terminated.")
```
